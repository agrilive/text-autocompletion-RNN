# Data Cleaning of email dataset

### Task

Prepare clean sentence tokens for training the autocompletion RNN model.

### Steps

1. Tokenize the message body into sentences with nltk's sentence tokenizer
2. Remove line breaks (e.g. \n, \t)
3. Remove URL and email addresses
4. Remove numbers
5. Expand contractions (e.g. don't becomes do not)
6. Remove punctuations

```python
import pandas as pd
file_name = 'enron_emails_clean_10k.csv'
df = pd.read_csv(file_name)
df = df.fillna('')
print(df.shape)
df.head()
```

```python
import re
from nltk.tokenize import sent_tokenize
sent_list = []
for idx, row in df[:2500].iterrows(): # 2500:7500 48630rows
    sentences = sent_tokenize(re.sub('\n|\r|\t', ' ', row['msg_body']))
    for sent in sentences:
        sent_list.append(sent)
sent_df = pd.DataFrame()
sent_df['msg_sentence'] = sent_list
sent_df
```

```python
import string
def preprocess_text(text):
    text = str(text.lower())
    rm_line_breaks = re.sub('\n|\r|\t', ' ', text)
    rm_url = re.sub(r'http\S+', '', rm_line_breaks)
    rm_email = re.sub(r'[\w\.-]+@[\w\.-]+', '', rm_url)
    rm_digits = ' '.join(s for s in rm_email.split() if not any(c.isdigit() for c in s))
    rm_digits = rm_digits.replace("’", "'")
    rm_contractions = expand_contractions(rm_digits)
    words = rm_contractions.split()
    rm_punc = re.compile('[%s]' % re.escape('''!()[]{};:'"|\,<>/?@#$%^&*~'''))
    wordlist = [rm_punc.sub('', w) for w in words]
    words = " ".join(wordlist)
    rm_dash = re.sub('-', '_', words)
    rm_fullstops = re.sub('\W', ' ', rm_dash)
    wordlist = [word.strip(string.punctuation) for word in rm_fullstops.split(" ")]
    words = " ".join(wordlist)
    return re.sub(' +', ' ', words.lstrip().rstrip())
```

```python
from tqdm import tqdm
for i in tqdm(range(10000)):
    sent_df['msg_clean']=sent_df['msg_sentence'].map(lambda s:preprocess_text(s)) 
```
